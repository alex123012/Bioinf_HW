```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro, ttest_ind, ttest_rel
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar12/breast_cancer_1000_genes.tsv',
                 index_col=0, sep='\t')
df
```

```python
res = pd.DataFrame({}, columns=df.columns)
for i, j in df.iterrows():
    # Можно брать полное количество сэмплов, тк их всего 100
    if shapiro(j)[1] > 0.05:
        res = res.append(j)
print("Нормально распределнных генов -", res.shape[0])
```

```python
df = pd.read_csv('https://github.com/s-a-nersisyan/HSE_bioinformatics_2021/blob/master/seminar12/colon_cancer_tumor_vs_normal_paired_FPKM.tsv?raw=true',
                 sep='\t',) #index_col=0)
df['ttest_ind'] = 0
df['ttest_rel'] = 0
for i, j in df.iterrows():
#     print(i)
    df.iloc[i, -2] = ttest_ind(df.iloc[i, 1:6], df.iloc[i, 6:11])[1]
    df.iloc[i, -1] = ttest_rel(df.iloc[i, 1:6], df.iloc[i, 6:11])[1]

df.set_index('Unnamed: 0', inplace=True)
df['ind_diff'] = df.ttest_ind < 0.05
df['rel_diff'] = df.ttest_rel < 0.05
df['both_diff'] = df.ind_diff & df.rel_diff
df[df.both_diff]
```

```python
ind = set(df.sort_values('ttest_ind', ascending=True).iloc[:10].index)
rel = set(df.sort_values('ttest_rel', ascending=True).iloc[:10].index)
res1 = df.loc[ind & rel]
res1
```

```python
res2 = df.nsmallest(10, ['ttest_ind', 'ttest_rel'], keep='all')
res2
```
