```python
fig, ax = plt.subplots(2, 2, figsize=(20, 20))
def print_ax(n, color, ax, size=1):
    rv = t(size)
    df = pd.DataFrame()
    for i in range(1000):
        df[i] = rv.rvs(size=n)
    n_means = pd.Series(df.mean())
    sns.histplot(data=n_means, binrange=(-10, 10), bins=100, stat='density', color=color, ax=ax)
    return n_means

rv = t(1)
listax = [(0, 0), (0, 1), (1, 0), (1, 1)]
colors = ['blue', 'pink', 'orange', 'red']
ns = [2, 5, 14, 100]
for n, color, j in zip(ns, colors, listax):
    print_ax(n, color, ax[j])
    x = np.linspace(-10,10,100)
    sns.lineplot(x=x, y=rv.pdf(x), ax=ax[j])
```
